Excercise: Guess the number

```python
from random import randint
cpu_num = randint(1, 100)

print(f'cheatcode: {cpu_num}')

def guess_the_number():
    try:
        user_num = int(input("Guess the number b/w 1 to 100"))    
        if user_num not in range(1, 101):
            print("The number is not within range 1 to 100!")
        elif user_num > cpu_num:
            print("The number is greater, please enter lower number!")
        elif user_num < cpu_num:
            print("The number is lesser, please enter greater number!")
        else:
            print("The guessed number is correct, You won!")
            return True
        return False
    except ValueError:
        print("Please enter digits not characters!")


while True:
    status = guess_the_number()
    if status:
        break
```
