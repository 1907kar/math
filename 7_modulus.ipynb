##### Modulus
 - It is the remainder from the division operation

```python
a = 10
b = 3
```

```python
division = a/b
print(division)
```

```python
remainder = a%b
print(remainder)
```

##### Exercise
 - Loop through -5 to +5 and find whether the number is odd or even

```python
for n in range(-5, 6):
    if n%2  == 0:
        print(f'{n} is an even number')
    else:
        print(f'{n} is an odd number')
```
