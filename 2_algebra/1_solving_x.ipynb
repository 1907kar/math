##### Solving x using sympy, always ensure the equation is homogenious
     2x + 4 = 9 (equation)
     2x + 4 - 9 = 0 (This is homogenious, bring all to one side)

```python
import sympy as sp
from IPython.display import display, Math
```

```python
x = sp.symbols('x')
expr = 2*x + 4 - 9
sp.solve(expr)
```

```python
# one way, 
display(Math("\\text{The solution to the equation %s is %s}" %(sp.latex(expr), sp.solve(expr))))
```

```python
# the other way
display(Math("\\text{The solution to the equation } %s \\text{ is } %s" %(expr, sp.solve(expr))))
```

```python
# completely solving the equation. sp.solve always puts in a list, so get its index, use %g
display(Math("\\text{The solution to the equation %s is %g}" %(sp.latex(expr), sp.solve(expr)[0])))
```

```python
# How to solve when there are two variables

y = sp.symbols('y')
expr = x/4 - x*y + 5
sp.solve(expr)
```

```python
# solve specific to x
sp.solve(expr, x)
```

```python
# solve specific to y
sp.solve(expr, y)
```
