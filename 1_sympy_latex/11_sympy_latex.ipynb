```python
import sympy as sp
from IPython.display import display, Math
```

```python
mu,alpha,sigma = sp.symbols('mu, alpha, sigma')
expr = sp.exp((mu - alpha)**2 / (2 * sigma**2))
display(expr)
```

```python
hello = sp.symbols('hello')
hello/3
```

```python
x,y = sp.symbols('x y')
expr = x+4+2*y
display(expr)
expr.subs({x:-4,y:3})
```

```python
x = sp.symbols('x')
expr = x**2 + 4

for n in range(-2, 3):
    display(Math(sp.latex(expr) + " = " + sp.latex(expr.subs(x, n))))
```
