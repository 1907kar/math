Avoid using fstring mix with latex as it make implementation and readable bit difficult


```python
from IPython.display import display, Math
import sympy as sp
```

```python
svar = 'Mike'
nvar = 36

print(f"The associate name is {svar} and the age is {nvar}.")
```

```python
x,y = sp.symbols('x y')

display(Math(f'\\frac{x}{y}'))

# we need to use double {{}} to display the fraction in f-string
display(Math(f'\\frac{{x}}{{y}}'))

# if we add latex expr then we need to extend the {}
expr = x/(3*y)

display(Math(f'\\frac{{{sp.latex(expr)}}}{{y}}'))

# instead of using f-string the better way to read is to use string formatting 
display(Math('\\frac{%s}{%s}' % (sp.latex(expr), y)))
```
