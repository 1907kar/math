##### Absolute value of a number
 - Its a distance value from its origin 0
 
    ##### Notation
      - abs(n) or |n|
                                                                                                                                                                                                                                                                                                                                                                

```python
from IPython.display import display, Math
```

```python
x = -4.3

display(Math('|%g| = %g' %(x, abs(x))))
```

##### Exercise

```python
num = {-4, 6, -1, 43, -18, 2, 0}
for n in num:
    if n < -5 or n > 2:
        print(f'Absolute value of {n} is {abs(n)}.')
    else:
        print(f'{n} was not tested.')
```
