##### Law of exponents
 1. In multiplication, When the bases are same with different exponents then that can be added
    - for ex: see RULE -1
 2. In division, When the bases are same with different exponents then that can be subracted
    - for ex: see RULE -2
 3. When the bases are different, then the exponents can't be simplified further
    - for ex: see RULE -3

```python
import sympy as sp
from IPython.display import display, Math
```

```python
# __________________________________________________
# RULE -1
# __________________________________________________
x, y, z = sp.symbols('x y z')
expr = x**y * x**z
display(Math(sp.latex(expr)))
# if we use below initialization, then inserting Math, latex can be ignored for these kind of expressions
sp.init_printing()
display(expr)
# as per rule -1, the above expression can be simplified as below
display(sp.simplify(expr))
```

```python
# __________________________________________________
# RULE -2
# __________________________________________________
x, y, z = sp.symbols('x y z')
expr = x**y / x**z
sp.init_printing()
# if we use sp.latex, on above expression, it will be already ignore the fraction in form
display(Math(sp.latex(expr)))
# so for illustration purpose, we will use Math and original latex to show the fraction form of the expression
display(Math('\\frac{x^y}{x^z}'))
# as per rule -2, the above expression can be simplified as below
display(sp.simplify(expr))
```

```python
# __________________________________________________
# RULE -3
# __________________________________________________
x, y, z = sp.symbols('x y z')
expr = x**y * y**z
sp.init_printing()
display(expr)
# as per rule -3, the above expression can't be simplified further
display(sp.simplify(expr))
```

```python
# When Math is really required?, if there is an involvement of strings then math is required.
x, y, z = sp.symbols('x y z')
expr_1 = x**y * x**z
expr_2 = x**y / x**z
expr_3 = x**y * y**z
display(Math('%s = \\text{(also simplified as) } %s' %(sp.latex(expr_1), sp.latex(sp.simplify(expr_1)))))
display(Math('%s = \\text{(also simplified as) } %s' %('\\frac{x^y}{x^z}', sp.latex(sp.simplify(expr_2)))))
display(Math('%s = \\text{(cannot be simplified) } %s' %(sp.latex(expr_3), sp.latex(sp.simplify(expr_3)))))
```

```python
# Different version of expression simplification are
x, y, z = sp.symbols('x y z')
expr_1 = x**y * x**z
# 1
display(sp.Eq(expr_1, sp.simplify(expr_1)))
# 2
display(sp.powsimp(expr_1))
```
