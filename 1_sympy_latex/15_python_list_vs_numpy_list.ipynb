```python
import numpy as np
```

```python
py_list = [1, 2, 3]
sub_list = [4]

print(py_list * 3)
# print(py_list + 3) # This will not work as it will try to concatenate the list with an integer, which is not allowed in Python.
print(py_list + sub_list)
```

```python
np_list = np.array([1, 2, 3])
sub_list = np.array([4])
print(np_list * 3)
print(np_list + 3)
print(np_list + sub_list)
```

```python
li = [3, ['3', '4', '5'], 5.5, [4, 5, 6]]
```

```python
for n, item in enumerate(li):
    print(f"List element {n} is {item} and is of type {type(item)}")
```

```python
for n, item in enumerate(li):
    print(f"List element {n} is {item} and is of type {type(item).__name__}")
```

```python
# Other way to get the type of the item
for n, item in enumerate(li):
    print(f"List element {n} is {item} and is of type {str(type(item))[8:-2]}")
```
